# Partie 1 - Data Cleaning

Je garde le vrai code dans `src/data_cleaning_part1.py` pour eviter un projet trop lourd.

Dans ce notebook, je relance simplement les memes etapes une par une pour les suivre plus facilement.

```python
from pathlib import Path
import sys
import pandas as pd

# Je retrouve la racine du projet pour pouvoir importer le script Python proprement.
ROOT_DIR = Path.cwd().resolve()
if not (ROOT_DIR / "src" / "data_cleaning_part1.py").exists() and (ROOT_DIR.parent / "src" / "data_cleaning_part1.py").exists():
    ROOT_DIR = ROOT_DIR.parent

SRC_DIR = ROOT_DIR / "src"
if str(SRC_DIR) not in sys.path:
    sys.path.append(str(SRC_DIR))

from data_cleaning_part1 import (
    load_em_companies,
    keep_only_common_isins,
    build_monthly_data,
    build_annual_data,
    build_base_investment_set,
    save_outputs,
)

pd.set_option("display.max_columns", 20)
```

## Etape 1 - Charger les entreprises EM

Je pars du fichier statique et je garde uniquement les entreprises de la region `EM`.

```python
em_companies = load_em_companies()
em_companies.head()
```

## Etape 2 - Retirer les ISIN manquants dans une table utile

Je retire ici les entreprises dont la ligne complete manque dans au moins un fichier indispensable.

```python
em_companies = keep_only_common_isins(em_companies)
em_companies.head()
```

## Etape 3 - Nettoyer les donnees mensuelles

Je traite ici les prix trop faibles, les delistings et je calcule les rendements mensuels.

```python
monthly_data = build_monthly_data(em_companies)
monthly_data.head()
```

## Etape 4 - Nettoyer les donnees annuelles

Je prepare maintenant `Scope 1`, `Revenue` et les prix de fin d'annee utiles pour la suite.

```python
annual_data = build_annual_data(em_companies, monthly_data)
annual_data.head()
```

## Etape 5 - Construire l'univers investissable de base

Je calcule ici le filtre de stale prices et je prepare la table qui servira a la partie 2.1.

```python
base_investment_set = build_base_investment_set(monthly_data)
base_investment_set.head()
```

## Etape 6 - Enregistrer les sorties

J'ecris les fichiers Excel dans `data/processed`.

```python
written_files = save_outputs(
    em_companies=em_companies,
    monthly_data=monthly_data,
    annual_data=annual_data,
    base_investment_set=base_investment_set,
)

written_files
```

## Resume final

Je verifie rapidement que tout est bien parti dans les bonnes tables.

```python
print(f"Nombre d'entreprises EM retenues : {len(em_companies)}")
print(f"Lignes du fichier mensuel : {len(monthly_data)}")
print(f"Lignes du fichier annuel : {len(annual_data)}")
print(f"Lignes du base investment set : {len(base_investment_set)}")
```
